```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_number_of_schools = len(school_data_complete["school_name"].unique())
total_number_of_students = school_data_complete["student_name"].count()
average_math_score = round(school_data_complete['math_score'].mean(), 2)
average_reading_score = round(school_data_complete['reading_score'].mean(), 2)
overall_passing_rate = round((average_math_score + average_reading_score)/2, 2)

students_with_passing_math_score = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]["student_name"].count()
students_with_passing_reading_score = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]["student_name"].count()
school_budgets = school_data_complete["budget"].value_counts().index.tolist()
total_school_budget = 0
for x in school_budgets:
    total_school_budget += x
percent_students_passing_math_score = round(students_with_passing_math_score * 100/total_number_of_students, 2)
percent_students_passing_reading_score = round(students_with_passing_reading_score * 100/total_number_of_students, 2 )

print(f"Total number of schools: {total_number_of_schools}")
print(f"Total number of students: {total_number_of_students}")
print(f"Total budget: {total_school_budget}")
print(f"Average math score: {average_math_score}")
print(f"Average reading score: {average_reading_score}")
print(f"Overall passing rate: {overall_passing_rate}")
print(f"Percentage of students with a passing math score: {percent_students_passing_math_score}%")
print(f"Percentage of students with a passing reading score: {percent_students_passing_reading_score}%")
d = {
    'Number of Schools': [total_number_of_schools],
    'Number of Students': [total_number_of_students],
    'Total budget': [total_school_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    'Overall Passing Rate': [overall_passing_rate],
    '% Students with Passing Math Score': [percent_students_passing_math_score],
    '% Students with Passing Reading Score': [percent_students_passing_reading_score]
}    
pd.DataFrame(data=d)


```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_data_list = school_data_complete["school_name"].value_counts().index.tolist()
data = {}

for x in school_data_list:
    data[x] = {}
    temp = school_data_complete.loc[school_data_complete["school_name"] == x, :]
    data[x]["School Type"] = temp["type"].unique()[0]
    data[x]["Total Number of Students"] = temp["student_name"].count()
    data[x]["Total School Budget"] = temp["budget"].unique()[0]
    data[x]["Per Student Budget"] = data[x]["Total School Budget"] / data[x]["Total Number of Students"]
    data[x]["Average Math Score"] = round(temp["math_score"].mean(), 2)
    data[x]["Average Reading Score"] = round(temp["reading_score"].mean(), 2)
    students_with_passing_math_score = temp.loc[school_data_complete["math_score"] >= 70, :]["student_name"].count()
    students_with_passing_reading_score = temp.loc[school_data_complete["reading_score"] >= 70, :]["student_name"].count()

    data[x]["%Passing Math"] = round(students_with_passing_math_score * 100 / data[x]["Total Number of Students"], 2)
    data[x]["%Passing Reading"] = round(students_with_passing_reading_score * 100 / data[x]["Total Number of Students"], 2)
    data[x]["Overall Passing Rate"] = round((data[x]["%Passing Math"] + data[x]["%Passing Reading"] ) / 2, 2)
    
summary_by_school_name = pd.DataFrame(data).T
summary_by_school_name
```

## Math Scores by Grade

  * Combine the series into a dataframe

```python
data = {}
for x in school_data_list:
    temp = student_data.loc[student_data["school_name"] == x]
    data[x] = {}
    data[x]["9th"] = round(temp.loc[temp["grade"] == "9th"]["math_score"].mean(), 2)
    data[x]["10th"] = round(temp.loc[temp["grade"] == "10th"]["math_score"].mean(), 2)
    data[x]["11th"] = round(temp.loc[temp["grade"] == "11th"]["math_score"].mean(), 2)
    data[x]["12th"] = round(temp.loc[temp["grade"] == "12th"]["math_score"].mean(), 2)
pd.DataFrame(data).T[["9th", '10th', '11th', '12th']]
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
data = {}
for x in school_data_list:
    temp = student_data.loc[student_data["school_name"] == x]
    data[x] = {}
    data[x]["9th"] = round(temp.loc[temp["grade"] == "9th"]["reading_score"].mean(), 2)
    data[x]["10th"] = round(temp.loc[temp["grade"] == "10th"]["reading_score"].mean(), 2)
    data[x]["11th"] = round(temp.loc[temp["grade"] == "11th"]["reading_score"].mean(), 2)
    data[x]["12th"] = round(temp.loc[temp["grade"] == "12th"]["reading_score"].mean(), 2)
pd.DataFrame(data).T[["9th", '10th', '11th', '12th']]
```
